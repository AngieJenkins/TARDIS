```python
import pandas as pd
import collections
from collections import Counter
import emoji
from emoji import UNICODE_EMOJI
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
tweet_table = pd.read_csv('tweets.csv')
```

```python
def str_len(item):
    words = str(item).split()
    len_string.append(len(words))
    return len_string
```

```python
# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –∏–º–µ–Ω–∏ –∞–≤—Ç–æ—Ä–∞
len_string = []
tweet_table['author_name'].apply(str_len)
tweet_table['len_name'] = len_string
```

```python
# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã —Ç–≤–∏—Ç–∞
len_string = []
tweet_table['tweet_content'].apply(str_len)
tweet_table['len_tweet'] = len_string
```

```python
# –°–∞–º—ã–µ –∫–æ—Ä–æ—Ç–∫–∏–µ —Ç–≤–∏—Ç—ã –∫–∞–∂–¥–æ–≥–æ –∞–≤—Ç–æ—Ä–∞
tweet_table_1 = tweet_table[tweet_table['len_name'] > 1]
authors = tweet_table_1.groupby('author_name')

for one in authors:
    df = pd.DataFrame(one[1])
    df2 = df.sort_values(by='len_tweet').head(1)
    author = df2.iloc[0, 0]
    tweet = df2.iloc[0, 5]
    print('Author: ', author, '\n', tweet, '\n\n')
```

```python
def find_emoji(item):
    try:
        words = item.split()
    except AttributeError:
        words = ''   
        
    for word in words:
        for symb in word:
            if symb in emoji.UNICODE_EMOJI:
                emjs.append(symb)
```

```python
emjs = []
tweet_table['tweet_content'].apply(find_emoji)
pop_emjs = Counter(emjs).most_common()
print(pop_emjs)
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
# –ü–æ–¥–≥–æ—Ç–æ–≤–∏–º —Å–ø–∏—Å–∫–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ –∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ –æ–∫—Ä–∞—à–µ–Ω–Ω—ã—Ö —Å–ª–æ–≤

positive = {}
negative = {}

with open('Positive_words.txt', 'r', encoding = 'utf-8') as pos:
    for line in pos:
        word = line.strip()
        positive[word] = 0

with open('Negative_words.txt', 'r', encoding = 'utf-8') as neg:
    for line in neg:
        word = line.strip()
        negative[word] = 0
```

```python
# –ü—Ä–æ–ø–∏—à–µ–º –Ω—É–∂–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ 
```

```python
def lemmatize(item):
    word_list = nltk.word_tokenize(str(item))
    for w in word_list:
        w = lemmatizer.lemmatize(w)
    lemm_words.append(word_list)
```

```python
def sent_score(item):
    score = 0
    for w in item:
        w = w.lower()
        if w in positive:
            score += 1
        elif w in negative:
            score -= 1
        else:
            pass
        
    sentiment_score.append(score)
    
    if score > 0:
        for w in item:
            w = w.lower()
            if w in positive:
                positive[w] += 1 
    
    elif score < 0:
        for w in item:
            w = w.lower()
            if w in negative:
                negative[w] += 1

    else:
        pass
```

```python
lemm_words = []
tweet_table['tweet_content'].apply(lemmatize)
tweet_table['lemmatized_tweets'] = lemm_words
```

```python
sentiment_score = []
tweet_table['lemmatized_tweets'].apply(sent_score)
tweet_table['sentiment_score'] = sentiment_score
```

```python
# –ó–∞–¥–∞–Ω–∏–µ 3.1

tweet_table = tweet_table.sort_values('sentiment_score')
most_expressive = tweet_table[(tweet_table['sentiment_score'] <= -8) | (tweet_table['sentiment_score'] >= 8)]
print(most_expressive)
```

```python
# –ó–∞–¥–∞–Ω–∏–µ 3.2

positive_words = []
negative_words = []
positive_values = []
negative_values = []
positive_dict = {}
negative_dict = {}

for key, value in positive.items():
    positive_words.append(key)
    positive_values.append(value)

for key, value in negative.items():
    negative_words.append(key)
    negative_values.append(value)

positive_dict['words'] = positive_words
positive_dict['values'] = positive_values 
negative_dict['words'] = negative_words
negative_dict['values'] = negative_values 

positive_frame = pd.DataFrame(positive_dict)
negative_frame = pd.DataFrame(negative_dict)

positive_frame = positive_frame.sort_values('values')
negative_frame = negative_frame.sort_values('values')

print('Most popular positive words: \n', positive_frame.tail(10))
print('\nMost popular negative words: \n', negative_frame.tail(10))
```
